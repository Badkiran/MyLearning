```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import  DecisionTreeClassifier
from sklearn import tree #For plotting a tree
from sklearn.metrics import classification_report 
from sklearn import preprocessing
```

```python
df = pd.read_csv("Company_Data.csv")
df
```

```python
df['Sale'] = np.where(df['Sales']> 8.0, 'High', 'Low')
df
```

```python
#Count of duplicated rows
df[df.duplicated()].shape
```

```python
data = df.iloc[:,1:]
data
```

```python
#label_encoder = preprocessing.LabelEncoder()
#company['Sale']= label_encoder.fit_transform(company['Sale']) 
#company
```

```python
company = pd.get_dummies(data.iloc[:,0:10])
company
```

```python
company['Sale'] = data['Sale'].map({'High':1 ,'Low':0})
company
```

```python
company.describe()company.describe()
```

```python
#Separating dependent and independent variables
X = company.iloc[:,0:-1]
y = company['Sale']
```

```python
X
```

```python
y
```

```python
X_train, X_test,y_train,y_test = train_test_split(X,y, test_size=0.2,random_state=10)
```

```python
model = DecisionTreeClassifier(criterion = 'entropy',max_depth=3)
model.fit(X_train,y_train)
```

```python
#PLot the decision tree
tree.plot_tree(model);
```

```python
fn=['CompPrice', 'Income','Advertising','Population','Price','Age','Education','ShelveLoc_Bad','ShelveLoc_Good','ShelveLoc_Medium','Urban_No','Urban_Yes','US_No','US_Yes']
cn=['High', 'Low']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=300)
tree.plot_tree(model,
               feature_names = fn, 
               class_names=cn,
               filled = True);
```

```python
preds = model.predict(X_test) 
pd.Series(preds).value_counts() # getting the count of each category 
```

```python
pd.crosstab(y_test,preds)
```

```python
# Accuracy 
np.mean(preds==y_test)
```

```python
actual_vs_predict = pd.DataFrame({'Actual': y_test,'Prediction': preds})
actual_vs_predict
```

```python
print(model.feature_importances_)
```

Good Shelf Location at store, Price, Advertising, Income and CompPrice are the leading features which generate high sale

```python
importance = model.feature_importances_
for i,v in enumerate(importance):
	print('Feature: %0d, Score: %.5f' % (i,v))
#plot feature importance
plt.bar([x for x in range(len(importance))], importance)
plt.show()
```
